```python

import requests
import pandas as pd

# NVD API endpoint for retrieving CVSS scores for a CPE
nvd_api = 'https://services.nvd.nist.gov/rest/json/cves/1.0?cpeName={}'

# List of CPE strings
cpe_list = ['cpe:2.3:o:microsoft:windows_10:1803:*:*:*:*:*:*:*', 'cpe:2.3:o:microsoft:windows_10:1809:*:*:*:*:*:*:*']

# Function to retrieve CVSS scores for a CPE
def get_cves_for_cpe(cpe):
    response = requests.get(nvd_api.format(cpe))
    cves = []
    if response.status_code == 200:
        data = response.json()
        if 'result' in data:
            cves = [cve['cve']['CVE_data_meta']['ID'] for cve in data['result']['CVE_Items']]
    return cves

# Retrieve CVSS scores for each CPE in the list
cves_dict = {}
for cpe in cpe_list:
    cves = get_cves_for_cpe(cpe)
    cves_dict[cpe] = cves

# Determine the CPE with the largest number of CVES
max_cves = 0
for cpe, cves in cves_dict.items():
    if len(cves) > max_cves:
        max_cves = len(cves)

# Create a dataframe with two columns, labeled with the names of the two CPEs
df = pd.DataFrame(columns=[cpe_list[0], cpe_list[1]])

# Fill in the dataframe with the CVES for each CPE
for i in range(max_cves):
    row = [None, None]
    for j, cpe in enumerate(cpe_list):
        cves = cves_dict[cpe]
        if i < len(cves):
            row[j] = cves[i]
    df.loc[i] = row

print(df)
df.bottom(20)
```

```python
import requests

# NVD API endpoint for retrieving CVSS scores for a CPE
nvd_api = 'https://services.nvd.nist.gov/rest/json/cves/1.0?cpeName={}'

# List of CPE strings
cpe_list = ['cpe:2.3:o:microsoft:windows_10:1803:*:*:*:*:*:*:*', 'cpe:2.3:o:microsoft:windows_10:1809:*:*:*:*:*:*:*']

# Function to retrieve CVSS scores for a CPE
def get_cves_for_cpe(cpe):
    response = requests.get(nvd_api.format(cpe))
    cves = []
    if response.status_code == 200:
        data = response.json()
        if 'result' in data:
            cves = [cve['cve']['CVE_data_meta']['ID'] for cve in data['result']['CVE_Items']]
    return cves

# Retrieve CVSS scores for each CPE in the list
cves_dict = {}
for cpe in cpe_list:
    cves = get_cves_for_cpe(cpe)
    cves_dict[cpe] = cves

# Print the total number of CVES for each CPE
for cpe, cves in cves_dict.items():
    print(f'The total number of CVES for {cpe} is {len(cves)}')

# Compute the number of CVES that are common to both CPEs
cpes_cves_intersection = set(cves_dict[cpe_list[0]]) & set(cves_dict[cpe_list[1]])
print(f'The total number of common CVES is: {len(cpes_cves_intersection)}')
df.head(25)
```
